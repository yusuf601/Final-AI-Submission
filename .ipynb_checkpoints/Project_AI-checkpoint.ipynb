# dari EDA diatas dapat dikonfirmasi bahwa tipe missing value adalah Missing At Random(MAR)

```python
def get_missing_gaps(series):
    """
    Mengembalikan list panjang missing gap berturut-turut.
    """
    gaps = []
    count = 0
    
    for value in series.isnull():
        if value:
            count += 1
        else:
            if count > 0:
                gaps.append(count)
                count = 0
    
    if count > 0:  # handle gap di akhir
        gaps.append(count)
    
    return gaps
```

```python
df_clean = df.copy()   # df = dataset komoditas tertentu

threshold = 0.20
cols_to_drop = []

print("="*70)
print("DROP PROVINSI MISSING > 20%")
print("="*70)

for col in df_clean.columns:
    if col == "Date":
        continue
    
    missing_pct = df_clean[col].isnull().mean()

    if missing_pct > threshold:
        print(f"  ✗ {col:30s} | {missing_pct*100:5.2f}% dropped")
        cols_to_drop.append(col)

df_clean = df_clean.drop(columns=cols_to_drop)

print("\nSelesai drop! Total dropped:", len(cols_to_drop))
print("Provinsi tersisa:", len(df_clean.columns) - 1)
```
